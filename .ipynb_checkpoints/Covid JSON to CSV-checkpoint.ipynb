```python
# Memperoleh Informasi yang Lebih Lengkap

import numpy as np
import pandas as pd
cov_sumut = pd.DataFrame(cov_sumut_raw['list_perkembangan'])
print('Info cov_sumut:\n', cov_sumut.info())
print('\nLima data teratas cov_sumut:\n', cov_sumut.head())
```

```python
# Menjinakkan Data
cov_sumut_tidy = (cov_sumut.drop(columns=[item for item in cov_sumut.columns
if item.startswith('AKUMULASI')
or item.startswith('DIRAWAT')])
.rename(columns=str.lower)
.rename(columns={'kasus': 'kasus_baru'})
)
cov_sumut_tidy['tanggal'] = pd.to_datetime(cov_sumut_tidy['tanggal']*1e6, unit='ns')
print('Lima data teratas:\n', cov_sumut_tidy.head(30))
```

```python

# Apakah Pekan ini Lebih Baik?

cov_sumut_pekanan = (cov_sumut_tidy.set_index('tanggal')['kasus_baru']
.resample('W')
.sum()
.reset_index()
.rename(columns={'kasus_baru': 'jumlah'})
)
cov_sumut_pekanan['tahun'] = cov_sumut_pekanan['tanggal'].apply(lambda x: x.year)
cov_sumut_pekanan['pekan_ke'] = cov_sumut_pekanan['tanggal'].apply(lambda x: x.weekofyear)
cov_sumut_pekanan = cov_sumut_pekanan[['tahun', 'pekan_ke', 'jumlah']]

print('Info cov_sumut_pekanan:')
print(cov_sumut_pekanan.info())
print('\nLima data teratas cov_sumut_pekanan:\n', cov_sumut_pekanan.head())

cov_sumut_pekanan['jumlah_pekanlalu'] = cov_sumut_pekanan['jumlah'].shift().replace(np.nan, 0).astype(np.int)
cov_sumut_pekanan['lebih_baik'] = cov_sumut_pekanan['jumlah'] < cov_sumut_pekanan['jumlah_pekanlalu']

print('Sepuluh data teratas:\n', cov_sumut_pekanan.head(10))
```

```python

# Membuat Bar Chart

import matplotlib.pyplot as plt

plt.clf()
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(data=cov_sumut_pekanan, x='pekan_ke', height='jumlah',
color=['mediumseagreen' if x is True else 'salmon' for x in cov_sumut_pekanan['lebih_baik']])
fig.suptitle('Kasus Pekanan Positif COVID-19 di Sumut',
y=1.00, fontsize=16, fontweight='bold', ha='center')
ax.set_title('Kolom hijau menunjukan penambahan kasus baru lebih sedikit dibandingkan satu pekan sebelumnya',
fontsize=12)
ax.set_xlabel('')
ax.set_ylabel('Jumlah kasus')
ax.text(1, -0.1, 'Sumber data: covid.19.go.id', color='blue',
ha='right', transform=ax.transAxes)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python

# Pola dan Dinamika

cov_sumut_akumulasi = cov_sumut_tidy[['tanggal']].copy()
cov_sumut_akumulasi['akumulasi_aktif'] = (cov_sumut_tidy['kasus_baru'] - cov_sumut_tidy['sembuh'] - cov_sumut_tidy['meninggal']).cumsum()
cov_sumut_akumulasi['akumulasi_sembuh'] = cov_sumut_tidy['sembuh'].cumsum()
cov_sumut_akumulasi['akumulasi_meninggal'] = cov_sumut_tidy['meninggal'].cumsum()
cov_sumut_akumulasi.tail()
```

```python
# Membuat Line Chart


import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.clf()
fig, ax = plt.subplots(figsize=(10,5))
ax.plot('tanggal', 'akumulasi_aktif', data=cov_sumut_akumulasi, lw=2)

ax.set_title('Akumulasi aktif COVID-19 di Sumut',
fontsize=22)
ax.set_xlabel('')
ax.set_ylabel('Akumulasi aktif')
ax.text(1, -0.1, 'Sumber data: covid.19.go.id', color='blue',
ha='right', transform=ax.transAxes)

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))

plt.grid()
plt.tight_layout()
plt.show()
```
